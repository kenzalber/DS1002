10.  Load the `iris` dataset that we used in class.  Use `iris` as the name of your data frame.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

11.  What is the correlation between the variables `sepal_length` and `sepal_width` for all the irises?

```python
correlation = iris[['sepal_length','sepal_width']].corr()
print(correlation)
```

12.  Which of the three species has the highest correlation between `sepal_length` and `sepal_width`?  Show any calculations you use to support your answer.

```python
setosa_df = iris[iris['species'] == 'setosa']
corr_setosa = setosa_df[['sepal_length','sepal_width']].corr()
print(corr_setosa)

versicolor_df = iris[iris['species'] == 'versicolor']
corr_versicolor = versicolor_df[['sepal_length','sepal_width']].corr()
print(corr_versicolor)

virginica_df = iris[iris['species'] == 'virginica']
corr_virginica = virginica_df[['sepal_length','sepal_width']].corr()
print(corr_virginica)

#Setosa has the highest correlation between sepal_length and sepal_width
```

Determine the same thing but for each species separately:  for each species, which two variables have the most negative (or the least positive) correlation?  Again, include calculations to support your answer.
