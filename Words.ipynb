```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
```

```python
#Question One:
#Orginal data type of "words" = set
wordlist = list(words)
print("Length of word list = " + str(len(wordlist))) #Prints out length of worlist

#Question Two:
print(wordlist[:10])
#All of these words are fairly complex, and don't seem to follow a distinct pattern


#Question Three:
alpha_list = sorted(wordlist)
print(alpha_list[:10])


#Question Four:
#Counts how many words start with 'a'
count_a = 0
for n in alpha_list:
  if n[0] == 'a':
    count_a+=1
print(str(count_a) + " words start with 'a'")


#Question Five:
#Counts how many words contain a 'q'
count_q = 0
for n in alpha_list:
  if 'q' in n:
    count_q+=1
print(str(count_q) + " words contain 'q'")

#Determines what percent of words do not contain an 'e'
count_e = 0
for n in alpha_list:
  if 'e' in n:
    count_e+=1
no_e = len(alpha_list) - count_e
percent_no_e = (no_e / len(alpha_list)) * 100
print(str((round(percent_no_e))) + " percent of words do not contain an 'e'")


#Question Six:
import numpy as np
wordlist = np.array(wordlist)
alpha_list = np.array(alpha_list)
alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)] #creates a list containing every letter in the alphabet


#Question Seven:
import matplotlib.pyplot as plt

#Creates a list containing the counts of words starting with each letter
letter_counts = np.zeros(26)
for n in range(26):
  for i in alpha_list:
    if alphabet[n] == i[0]:
     letter_counts[n]+=1

#Creates bar graph of the number of words starting with each letter
plt.bar(alphabet, letter_counts, color='g')
plt.ylabel("Number of words")
plt.xlabel("Starting letter of word")
plt.title("Number of words starting with each letter")
plt.show()


#Question Eight:
#Six longest words in worst list
sorted_list = sorted(alpha_list, key=lambda x: len(x), reverse=True)
print(sorted_list[:6])

#Question Nine:
lengths = [len(word) for word in sorted_list]

plt.hist(lengths, bins=25)
plt.title("Histogram")
plt.show()

#Question Eleven:
v_and_a = [word for word in sorted_list if 'v' in word and 'a' in word]
w_and_l = [word for word in sorted_list if 'w' in word and 'l' in word]
print("Number of words that contain 'v' and 'a': " + str(len(v_and_a)))
print("Number of words that contain 'w' and 'l': " + str(len(w_and_l)))
```
